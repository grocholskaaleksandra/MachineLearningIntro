```python
import altair as alt #for interactive charts - to be used by non-technical people
from vega_datasets import data
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
source = data.cars()
source.info()
alt.Chart(source).mark_circle(size=60).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip=["Name", "Horsepower", "Miles_per_Gallon"] #displays additionnal data while hovering the mouse over a point
).interactive() #chart is movable, we can zoom in and out
```

```python
#On-class exercise
#1. Na podstawie dataset'u Penguins (wybierz odpowiednie kolumny) utwórz wykres (link)
#https://altair-viz.github.io/gallery/poly_fit_regression.html
#oraz:
#dodaj tooltip
#dodaj interacitve
#zapisz do html

penguins = sns.load_dataset("penguins")
penguins.info()
```

```python
#base chart

alt.Chart(penguins).mark_circle(size=60).encode(
    x='body_mass_g',
    y='bill_length_mm',
    color='sex',
    tooltip=["species", "island"]
).interactive()
```

```python
degree_list = [1, 3, 5] # Define the degree of the polynomial fits

base = alt.Chart(penguins).mark_circle(color="black").encode(
        alt.X("body_mass_g"), alt.Y("bill_length_mm"), 
        tooltip=["species", "island"]
) #base chart based on above one

polynomial_fit = [
    base.transform_regression( # https://altair-viz.github.io/user_guide/transform/regression.html
        "body_mass_g", "bill_length_mm", method="poly", order=order, as_=["body_mass_g", str(order)]
    )
    .mark_line() #A line plot.
    .transform_fold([str(order)], as_=["degree", "bill_length_mm"]) # https://altair-viz.github.io/user_guide/transform/fold.html
    .encode(alt.Color("degree:N"))
    for order in degree_list
] #polynommial fit

alt.layer(base, *polynomial_fit).interactive() #merge of base and polynomial fit
```
