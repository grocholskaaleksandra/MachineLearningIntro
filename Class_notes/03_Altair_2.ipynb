```python
import pandas as pd

dane = pd.read_csv('CSV.csv')
```

```python
print (dane)
```

```python
test = dane["Adj Close"].rolling(2).mean()
print(test)
```

```python
test.plot()
```

```python
import altair as alt

line = alt.Chart(dane).mark_line(
    color='red',
    size=3
).transform_window(
    rolling_mean='mean(Adj Close)',
    frame=[-15, 15]
).encode(
    x='Date:T',
    y='rolling_mean:Q'
)


points = alt.Chart(dane).mark_point().encode(
    x='Date:T',
    y=alt.Y('Adj Close:Q',
            axis=alt.Axis(title='Adj Close'))
)

alt.layer(points + line).interactive()
```
