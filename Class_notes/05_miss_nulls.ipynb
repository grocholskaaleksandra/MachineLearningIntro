```python
import pandas as pd
import seaborn as sns
import numpy as np

houses = pd.read_csv('houses_data.csv')
houses.dtypes
houses_numeric = houses.select_dtypes(include='number')
houses_numeric.dtypes

houses_numeric.isnull().values.any()
houses_numeric.isnull().sum()

```

```python
houses_na = houses[['Car', 'BuildingArea', 'YearBuilt']].copy()
houses_na2= houses[['Car', 'BuildingArea', 'YearBuilt', 'Price']].copy()
```

```python
houses_na
display(houses_na.describe()) 
```

```python
houses_new = houses_na.fillna(houses_na.mean())
display(houses_new.describe()) 
```

```python
houses_new2 = houses_na.fillna(method = 'bfill')
display(houses_new2.describe()) 
```

```python
houses_dropna = houses_na2.dropna(axis=0)
display(houses_na2.describe()) 
houses_dropna2 = houses_dropna[['Car', 'BuildingArea', 'YearBuilt']].copy()
display(houses_dropna2.describe()) 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(houses_dropna2,  
                                                    houses_dropna['Price'], #tutaj zmienic na kolumne cena
                                                    train_size=0.7,
                                                    test_size=0.3,
                                                    random_state=0)

def score_dataset(X_train, X_test, y_train, y_test):
    regr_model = LinearRegression()
    regr_model.fit(X_train, y_train)
    preds = regr_model.predict(X_test)
    return mean_absolute_error(y_test, preds)

score_dataset(X_train, X_test, y_train, y_test)
```
