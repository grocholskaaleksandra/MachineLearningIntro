```python
#Dataset:
# x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
# y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
# y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
# y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
# x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
# y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
# Calculate:
# mean,
# standard deviation,
# correlation coefficient (Pearson)
# variance 


x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
import numpy as np

def mean(dataset):
    mean = np.mean(dataset)
    print (mean)

def std(dataset):
    std = np.std(dataset)
    print (std)
    
def corr(dataset1, dataset2):
    corr = np.corrcoef(dataset1, dataset2)
    print (corr)
    
def var(dataset):
    var = np.var(dataset)
    print (var)
    
print ("Mean: ")
mean(x)
mean(y1)
mean(y2)
mean(y3)
mean(x4)
mean(y4)


print(" ")
print ("Standard deviation: ")
std(x)
std(y1)
std(y2)
std(y3)
std(x4)
std(y4)

print(" ")
print ("Correlation: ")
corr(x, y1)
corr(x, y2) 
corr(x, y3)
corr(x, y4)
corr(x, x4)
corr(y1, y2)
corr(y1, y3)
corr(y1, y4)
corr(y1, x4)
corr(y2, y3)
corr(y2, y4)
corr(y2, x4)
corr(y3, y4)
corr(y3, x4)
corr(y4, x4)

print(" ")
print ("Variation: ")
var(x)
var(y1)
var(y2)
var(y3)
var(x4)
var(y4)
```
