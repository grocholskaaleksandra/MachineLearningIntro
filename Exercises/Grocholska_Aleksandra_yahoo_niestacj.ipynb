```python
display (yahoo_finance)
```

```python
adj_close = yahoo_finance["Adj Close"] 
display(adj_close)
```

```python
adj_close.name = 'Adj close'
adj_close_50ma = adj_close.rolling(50).mean() #50 day moving average
adj_close_50ma.name = '50 day MA'
adj_close_200ma = adj_close.rolling(200).mean() #200 day moving average
adj_close_200ma.name = '200 day MA'
adj_close.plot(title='Yahoo finance', legend=True);adj_close_50ma.plot(legend=True);adj_close_200ma.plot(legend=True) 
```

```python
import altair as alt

line = alt.Chart(yahoo_finance).mark_line( 
    color='red',
    size=3
).transform_window(
    rolling_mean='mean(Adj Close)', 
    frame=[-20, 20]
).encode( 
    x='Date:T',
    y='rolling_mean:Q'
)


points = alt.Chart(yahoo_finance).mark_point().encode(
    x='Date:T',
    y=alt.Y('Adj Close:Q',
            axis=alt.Axis(title='Adj Close'))
)

alt.layer(points + line).interactive()
```
